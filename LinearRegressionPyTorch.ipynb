```python
import torch.nn as nn
import torch.nn.functional as F
# almost everything from PyTorch inherits from nn.Module
```

```python
class LinearRegressionModel(nn.Module): 
    def __init__(self):
        super().__init__() # you are essentially calling the constructor of the parent class (nn.Module) to ensure that any necessary setup or initialization from the base class is performed before you add your own custom parameters and functionality in the __init__ method of LinearRegressionModel
        
        # Initializing Model Parameters
        self.w = nn.Parameter(torch.tensor(0.0, requires_grad=True, dtype=torch.float64)) # creates parameter that will be automatically added to the list of model parameters
        self.b = nn.Parameter(torch.tensor(0.0, requires_grad=True, dtype=torch.float64))
    
    #defines forward method for prediction of model
    def forward(self, x):
        y = self.w * x + self.b
        return y
```

```python
import torch.optim
```

```python
model = LinearRegressionModel()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X = torch.tensor(data['Cost'].to_numpy())
Y = torch.tensor(data['Score'].to_numpy())
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, shuffle=True) 
```

```python
lr = 0.001
epochs = 500
```

```python
optim = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
train_loss_accum = np.array([])
test_loss_accum = np.array([])
w_accum = np.array([])
b_accum = np.array([])
```

```python
for i in range(epochs):
    # forward pass
    model.train()
    prediction = model.forward(X_train)
    
    # training computing the loss
    train_loss = ((Y_train - prediction)**2).sum()/(2*len(Y_train))
    
    #compute the new gradients
    optim.zero_grad() # resets gradient calculations from previous iterations
    train_loss.backward() # performs backword propagation (computes the graduients using a computation graph) 
    optim.step() # uses gradient desent to update model parameters
    
    # Testing the generalizability 
    model.eval() # turns off settings not needed for testing the model
    with torch.inference_mode():
        preds = model.forward(X_test)
        test_loss = ((Y_test - preds)**2).sum()/(2*len(Y_test))
    
    
    # loss accumulators / parameter accumulators
    train_loss_accum = np.append(train_loss_accum, train_loss.detach().numpy())
    test_loss_accum = np.append(test_loss_accum, test_loss.detach().numpy())
    w_accum = np.append(w_accum, model.w.detach().numpy())
    b_accum = np.append(b_accum, model.b.detach().numpy())
    
    
    print(f"Epoch: {i+1} | Train Loss: {train_loss} | Test Loss: {test_loss}")
```

```python
fig, axs = plt.subplots(1,2)
axs[0].set_xlabel("Epoch")
axs[0].set_ylabel("Train Loss")

axs[1].set_xlabel("Epoch")
axs[1].set_ylabel("Test Loss")

plt.subplots_adjust(wspace=0.4)

sns.lineplot(data=train_loss_accum, ax=axs[0]);
sns.lineplot(data=test_loss_accum, ax=axs[1]);
```
